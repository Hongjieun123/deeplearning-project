```
import argparse
import io

#구글에서 제공하는 튜토리얼
# https://github.com/googleapis/python-speech/blob/master/samples/v1/speech_transcribe_sync.py (원본)
# [START speech_transcribe_sync]
def transcribe_file(speech_file): #파일명 넣는다.
    """Transcribe the given audio file."""
    from google.cloud import speech
    from google.cloud.speech import enums
    from google.cloud.speech import types
    client = speech.SpeechClient()  #클라이언트 서버이기 때문에 클라이언트 만들어야, 2가지 있어야 처리 가능

    # [START speech_python_migration_sync_request]
    # [START speech_python_migration_config]
    with io.open(speech_file, 'rb') as audio_file: #오디오 파일 읽는다.
        content = audio_file.read()

    audio = types.RecognitionAudio(content=content)
    config = types.RecognitionConfig(
        encoding=enums.RecognitionConfig.AudioEncoding.LINEAR16, #오디오 파일 형식 (WAV=LINEAR16)
        # encoding=enums.RecognitionConfig.AudioEncoding.FLAC,  # FLAC 얘도 오디오 파일 형식
        sample_rate_hertz=16000,  #이미지안에 0~255 내부 숫자 들어 있음, 오디오는 시간 개념이므로 샘플 = 1초를 몇개의 샘플로 표현하겠다!!! 
        language_code='ko-KR') #language 마다 음성인식기가 따로 있다. 
        # language_code='en-US')
    # [END speech_python_migration_config]

    # operation = client.long_running_recognize(config, audio)
    #
    # print('Waiting for operation to complete...')
    # response = operation.result(timeout=90)

    # [START speech_python_migration_sync_response]
    response = client.recognize(config, audio)  #클라이언트는 CONFIG, AUDIO 파일 있으면 된다. 
    # dict 타입 이다. response는
    # [END speech_python_migration_sync_request]
    # Each result is for a consecutive portion of the audio. Iterate through
    # them to get the transcripts for the entire audio file.
    for result in response.results:
        # The first alternative is the most likely one for this portion.
        print(response.results)
        print(u'Transcript: {}'.format(result.alternatives[0].transcript))
        print(u'Confidence: {}'.format(result.alternatives[0].confidence))
    # [END speech_python_migration_sync_response]
# [END speech_transcribe_sync]


# [START speech_transcribe_sync_gcs]
def transcribe_gcs(gcs_uri): #1분이 넘어갈 때 쓴다
    """Transcribes the audio file specified by the gcs_uri."""
    from google.cloud import speech
    from google.cloud.speech import enums
    from google.cloud.speech import types
    client = speech.SpeechClient()

    # [START speech_python_migration_config_gcs]
    audio = types.RecognitionAudio(uri=gcs_uri)
    config = types.RecognitionConfig(
        encoding=enums.RecognitionConfig.AudioEncoding.FLAC,
        sample_rate_hertz=16000,
        language_code='en-US')
    # [END speech_python_migration_config_gcs]

    response = client.recognize(config, audio)
    # Each result is for a consecutive portion of the audio. Iterate through
    # them to get the transcripts for the entire audio file.
    for result in response.results:
        # The first alternative is the most likely one for this portion.
        print(u'Transcript: {}'.format(result.alternatives[0].transcript))
# [END speech_transcribe_sync_gcs]
```
