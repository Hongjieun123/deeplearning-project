```
import tensorflow as tf
```

```
a = tf.constant(3.0)   #노드 정의
b = tf.constant(4.0)
c = tf.constant(5.0)
d = a * b + c  #엣지 정의
print(d)  #아직 연산 안됨  d라는 노드 자체를 출력

sess = tf.Session()  #flow 단계
result = sess.run(d)
print(result)
```

image.png

```
W = tf.Variable(tf.random_normal(shape=[1], name='w'))
b = tf.Variable(tf.random_normal(shape=[1], name='b'))
x = tf.placeholder(tf.float32)

pred = W*x + b
```

image.png

```
y = tf.placeholder(tf.float32)

loss = tf.reduce_mean(tf.square(pred - y))
```

image.png

```
optimizer = tf.train.GradientDescentOptimizer(0.01)

train_step = optimizer.minimize(loss)  #loss를 최소화하도록 업데이트
```

```
x_train = [1,2,3,4]
y_train = [3,5,7,9]
```

```
sess = tf.Session()
result = sess.run(tf.global_variables_initializer())
```

```
for i in range(1000):
  sess.run(train_step, feed_dict={x: x_train, y:y_train})
```

```
sess.run([W,b])
```
